# Filters

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Filter parameters
R = 0.150         # [Ohm]
C = 10 * 10**(-6) # [F]

print(f"f_3dB = {1 / (2 * np.pi * R * C)} Hz")
print(f"Power rating R = {0.5**2 * R} W")
```

```python
# Frequency range
freq = np.logspace(0, 7, 500) # From 1 Hz to 100000 Hz
omega = 2 * np.pi * freq

# Transfer function
# H(jw) = 1 / (1 + jwRC)
transfer_function = 1 / (1 + 1j * omega * R * C)

# Calculate magnitude and phase
magnitude = np.abs(transfer_function)
magnitude_db = 20 * np.log10(magnitude)
phase_deg = np.rad2deg(np.angle(transfer_function))

# Create the Bode plot
plt.figure(figsize=(10, 6))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.semilogx(freq, magnitude_db, '-b')
plt.ylabel('Magnitude [dB]')
plt.title('Bode Plot of RC Circuit')
plt.grid(which='both', linestyle='--', linewidth=0.5)

# Phase plot
plt.subplot(2, 1, 2)
plt.semilogx(freq, phase_deg, '-g')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase []')
plt.grid(which='both', linestyle='--', linewidth=0.5)
```
