# USB Controlled Impedance
Requirements for USB 2.0 data lines:
- Differential impedance: 90 Ohm (+-15%)
- Common-mode impedance: 30 Ohm
- Similar lengths

```python
Z_diff = 90
Z_comm = 30

# Odd and Even impedance for KiCad
Z_odd = Z_diff / 2
Z_even = Z_comm * 2

print(f"{Z_odd=}")
print(f"{Z_even=}")
```

**Single-ended impedance** \
Z_{0} = sqrt(L_{0} / C_{0})

**Differential impedance** \
Z_{d} = sqrt((L_{0} - L_{m} / (C_{0} + 2 * C_{m})))

Some rules of thumb:
* Wider tracks -> Lower single ended impedance
* Tracks closer together -> Lower differential impedance
* Start with single-ended impedance to determine track width, followed by differential impedance to determine gap between tracks
